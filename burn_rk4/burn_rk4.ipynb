```python
import sys
sys.path.append(".")

import burn_rk4
import matplotlib.pyplot as plt
import numpy as np


steps = 1000
dt = 0.1
positions, velocities = burn_rk4.solve_oscillator(steps, dt)
time = np.linspace(0, steps * dt, steps)

plt.figure(figsize=(10, 5))

# Position vs Time
plt.subplot(1, 2, 1)
plt.plot(time, positions, label='Position', color='blue')
plt.title("Harmonic Oscillator (Burn RK4)")
plt.xlabel("Time")
plt.ylabel("Displacement")
plt.grid(True)

# Phase Portrait (Position vs Velocity)
plt.subplot(1, 2, 2)
plt.plot(positions, velocities, color='purple')
plt.title("Phase Space")
plt.xlabel("Position")
plt.ylabel("Velocity")
plt.grid(True)

plt.tight_layout()
plt.show()
```
